```python
import os
from pathlib import Path
```

```python
filename="en-training.txt"
ODir="./Training"
with open(filename) as data:
    p=Path(ODir)
    if not p.exists():
        os.mkdir(ODir)
    for feedback in data:
        feedback = feedback.split()
        fid=int(feedback[0].split('-')[2]) #en-training-0123
        del(feedback[0])
        dataclass=feedback[-1]
        p=Path(ODir+"/"+dataclass)
        if not p.exists():
            os.mkdir(ODir+"/"+dataclass)
        del(feedback[-1])
        feedbackVal=""
        for feeds in feedback:
            feedbackVal +=feeds
            feedbackVal +=" "
            fi=open(ODir+"/"+dataclass+"/"+str(fid),"w")
            fi.write(feedbackVal)
            fi.close()
        print(fid,feedbackVal,dataclass)
```

```python
import numpy as np
```

```python
path="./Training/"
p=Path(path)
subdir=os.listdir(path)
subdir
```

```python
data=[]
target=[]
for sdir in subdir:
    spath=path+sdir+"/"
#     print(spath)
    files=os.listdir(spath)
#     print(files)
    for file in files:
        filename=spath+file
#         print(filename)
        with open(spath+file) as content:
            data.append(content.read().strip().split())
            target.append(sdir)
#             print(content.read(),sdir)
print(len(data),len(target))

    
        

    
```

```python
print(data[:2],target[:2])
```

```python
#Getting stopword

stopwords = []

def get_the_sw():
    with open('stopwords.txt') as stops:
        for i in stops:
            stopwords.append(i.strip())
get_the_sw()
stopwords
```

```python
feedbacks = []

def get_the_corpus():
    for i in data:
        feedback = i
        feedback = [word for word in feedback if word not in stopwords ]
        feedbacks.append(feedback)
get_the_corpus()
feedbacks
```

```python
from collections import defaultdict
frequency = defaultdict(int)
def removal_singal_freq():
    for feedback in feedbacks:
        for word in feedback:
            frequency[word] += 1
removal_singal_freq()
feedbacks = [[word for word in feedback if frequency[word]>1] for feedback in feedbacks]
feedbacks
```

```python
import gensim
dictionary = gensim.corpora.Dictionary(feedbacks)
bag_of_words = [ dictionary.doc2bow(feedback) for feedback in feedbacks]
n_unique_tokens = len(dictionary)
dense_bow = gensim.matutils.corpus2dense(bag_of_words, num_terms = n_unique_tokens).transpose()
datset = dense_bow
```

```python
datset
```

```python
targt = []
for i in target:
    if i == 'bug':
        targt.append(1)
    elif i == 'comment':
        targt.append(2)
    elif i == 'complaint':
        targt.append(3)
    elif i == 'meaningless':
        targt.append(4)
    elif i == 'request':
        targt.append(5)
    elif i == 'undefined':
        targt.append(6)
targt
```

```python
print(len(datset),len(targt))
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=10,shuffle=True)
```

```python
from sklearn import svm
from sklearn import naive_bayes as nb
from sklearn.metrics import f1_score
from numpy import mean
accuracies = []
scores = []
```

```python
for train, test in kf.split(datset):
    print("1")
    train_set = []
    train_labels = []
    test_set = []
    test_labels = []
    for i in train:
        train_set.append(datset[i])
        train_labels.append(targt[i])
    for i in test:
        test_set.append(datset[i])
        test_labels.append(targt[i])
    classifier = nb.GaussianNB()
#     classifier = svm.SVC()
    predicted = classifier.fit(train_set, train_labels).predict(test_set)
    score = f1_score(test_labels, predicted, average = 'weighted')
    scores.append(score)
    incorrect = (test_labels != predicted).sum()
    accuracy = (len(test_set) - incorrect) / len(test_set) * 100.
    accuracies.append(accuracy)
#     print("1")
print("Maximum accuracy attained ", max(accuracies))
print("f1score  ", scores[np.argmax(accuracies)])
print('\n')
```
